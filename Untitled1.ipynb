```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
event_file_path = 'save_file/log_file/debug/events.out.tfevents.1702812949.81ac1df76984.3006.0'
ea = event_accumulator.EventAccumulator(event_file_path)

```

```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt
# Path to your TensorBoard logs directory
log_dir = 'save_file/log_file/debug/events.out.tfevents.1702812949.81ac1df76984.3006.0'

# Create an EventAccumulator object
ea = EventAccumulator(log_dir)
ea.Reload()
# Get the list of available tags
tags = ea.Tags()
print("Available Tags:")
for tag_type in tags:
    print(f"{tag_type}: {tags[tag_type]}")
```

```python
# List of tag names
tag_names = ['loss/total', 'loss/loss_ce', 'loss/loss_point']

# Plotting for each tag
plt.figure(figsize=(10, 6))
for tag_name in tag_names:
    values = ea.Scalars(tag_name)
    steps = [event.step for event in values]
    values = [event.value for event in values]
    plt.plot(steps, values, linestyle='-', label=tag_name)

plt.title('Losses Over Steps')
plt.xlabel('Steps')
plt.ylabel('Loss Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
